## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data wrangling is the process of cleaning, transforming, and organizing raw data into
# a usable format for analysis or machine learning.
````

````python
# 1. Handle Outliers (Winsorization)
# •	Handled outliers using Winsorization by limiting extreme values to the 1st and 99th percentiles.
# • This minimizes the impact of outliers without deleting any data, resulting
#   in more reliable analysis.
# •	Columns affected: all numeric features (float64, int64).
# -------------------------------
def cap_outliers(col):
    q1 = df[col].quantile(0.01)
    q99 = df[col].quantile(0.99)
    df[col] = np.clip(df[col], q1, q99)

num_cols = df.select_dtypes(include=['float64','int64']).columns
for col in num_cols:
    cap_outliers(col)
````

````python
# -------------------------------
# 2. Feature Engineering
# 	Created Debt-to-Income Ratio:
# "Debt-to-Income Ratio"="Outstanding Debt" /"Annual Income"
# → Helps measure how much debt a customer carries relative to their income.

# 	Created EMI-to-Salary Ratio:
# "EMI-to-Salary Ratio"="Total EMI per month" /"Monthly Inhand Salary"
# → Indicates affordability of monthly loan repayments compared to salary.

# -------------------------------
df['Debt_to_Income_Ratio'] = df['Outstanding_Debt'] / df['Annual_Income']
df['EMI_to_Salary_Ratio'] = df['Total_EMI_per_month'] / df['Monthly_Inhand_Salary']
````

````python

# -------------------------------
# 3. Encode Categorical Variables
# •	Converted categorical variables into numeric codes using .astype('category').cat.codes.
# •	This makes the dataset machine-learning ready while preserving categorical distinctions.
# -------------------------------
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].astype('category').cat.codes
````

### What all manipulations have you done and insights you found?

1.  **Data Readiness**
    *   The dataset is now analysis-ready: no missing values, no extreme outliers, categorical variables encoded, and new ratios added for deeper insights.
2.  **Debt-to-Income Ratio**
    *   Customers with high debt relative to income are likely to have lower credit scores.
    *   This ratio is a strong indicator of financial stress and repayment risk.
3.  **EMI-to-Salary Ratio**
    *   A high EMI-to-Salary ratio suggests repayment burden is heavy compared to income.
    *   These customers are more vulnerable to defaults and delayed payments.
